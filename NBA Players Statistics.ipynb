### NBA Season Totals

```python
# load the necessary packages
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
# This imports the client
from basketball_reference_web_scraper import client
```

```python
#Get data for all NBA players
def load_data(year):
    data = client.players_season_totals(season_end_year=year)
    nba = pd.DataFrame(data=data, index=None)
    return nba
```

```python
#NBA players statistics per year
nba_data = load_data(2022)
nba_data.head(10)
```

```python
# Clean up Positions Column
positions_clean = nba_data['positions'].apply(lambda x: ''.join(map(str, x))).str.split(".", n = 1, expand = True)
nba_data['position'] = positions_clean[1]
nba_data.drop(columns =['positions'], inplace = True)
```

```python
# Clean up team Column
a=lambda x: str(x).split('.')
team_clean = nba_data['team'].apply(a)
team_clean = team_clean.apply(lambda x: ','.join(map(str, x))).str.split(",", n = 1, expand = True)
nba_data['nba_team'] = team_clean[1]
nba_data.drop(columns =['team'], inplace = True)
nba_data.head()
```

```python
# Duplicate rows due to this player changing teams
nba_data[nba_data['name'] == 'Gary Harris']
```

```python
#Format data using SQL
from pandasql import sqldf

query = """
        SELECT
            slug,
            name,
            position,
            SUM(games_played) AS games_played,
            SUM(games_started) AS games_started,
            SUM(minutes_played) AS minutes_played,
            SUM(made_field_goals) AS made_field_goals,
            SUM(attempted_field_goals) AS attempted_field_goals,
            SUM(made_three_point_field_goals) AS made_three_point_field_goals,
            SUM(attempted_three_point_field_goals) AS attempted_three_point_field_goals,
            SUM(made_free_throws) AS made_free_throws,
            SUM(attempted_free_throws) AS attempted_free_throws,
            SUM(offensive_rebounds) AS offensive_rebounds,
            SUM(defensive_rebounds) AS defensive_rebounds,
            SUM(assists) AS assists,
            SUM(steals) AS steals,
            SUM(blocks) AS blocks,
            SUM(turnovers) AS turnovers,
            SUM(personal_fouls) AS personal_fouls,
            SUM(points) AS points
        FROM nba_data
        GROUP BY 1,2,3

"""



nba_data_cleaned = sqldf(query, locals())
```

```python
nba_data_cleaned.head()
```

```python
# Add other statistical measures. ie: PPG, RPG
nba_data_cleaned['MPG'] = round(nba_data_cleaned['minutes_played']/nba_data_cleaned['games_played'],1)
nba_data_cleaned['PPG'] = round(nba_data_cleaned['points']/nba_data_cleaned['games_played'],1)
nba_data_cleaned['RPG'] = round((nba_data_cleaned['offensive_rebounds'] + nba_data_cleaned['defensive_rebounds'])/nba_data_cleaned['games_played'],1)
nba_data_cleaned['APG'] = round(nba_data_cleaned['assists']/nba_data_cleaned['games_played'],1)
nba_data_cleaned['FG%'] = round(nba_data_cleaned['made_field_goals']/nba_data_cleaned['attempted_field_goals'],3)
nba_data_cleaned['FT'] = round(nba_data_cleaned['made_free_throws']/nba_data_cleaned['games_played'],1)
nba_data_cleaned['FT%'] = round(nba_data_cleaned['made_free_throws']/nba_data_cleaned['attempted_free_throws'],3)
nba_data_cleaned['3P'] = round(nba_data_cleaned['made_three_point_field_goals']/nba_data_cleaned['games_played'],1)
nba_data_cleaned['3P%'] = round(nba_data_cleaned['made_three_point_field_goals']/nba_data_cleaned['attempted_three_point_field_goals'],1)
nba_data_cleaned['STPG'] = round(nba_data_cleaned['steals']/nba_data_cleaned['games_played'],1)
nba_data_cleaned['BLKPG'] = round(nba_data_cleaned['blocks']/nba_data_cleaned['games_played'],1)
nba_data_cleaned['TOPG'] = round(nba_data_cleaned['turnovers']/nba_data_cleaned['games_played'],1)
nba_data_cleaned = nba_data_cleaned.fillna(0)
```

```python
# New dataframe with most important information for Fantasy
nba_dataset = nba_data_cleaned[['name', 'games_played', 'PPG', 'FG%', 'FT', 'FT%', '3P', '3P%', 'RPG', 'APG', 'STPG', 'BLKPG', 'TOPG']]
nba_dataset.head()
```
